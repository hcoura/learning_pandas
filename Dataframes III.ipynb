# The .query() Method

```python
cols = bond.columns.map(lambda x: x.replace(' ', '_'))
bond.columns = cols
bond.head(3)
```

```python
bond.query('Actor == "Sean Connery"').head(3)
```

```python
a = "Sean Connery"
bond.query('Actor == @a').head(3)
```
