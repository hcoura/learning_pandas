# Extract Rows from MultiIndexes

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], 
                     index_col=["Date", "Country"]).sort_index()
bigmac.head(3)
```

```python
bigmac.loc[("2010-01-01", "Brazil")]
```

```python
bigmac.loc[("2010-01-01", "Brazil"), "Price in US Dollars"]
```

```python
bigmac.loc[("2015-07-01",)].head()
```

# .transpose()

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates=["Date"], 
                     index_col=["Date", "Country"]).sort_index()
bigmac.head(3)
```

```python
bigmac = bigmac.transpose()
bigmac.head(3)
```

```python
bigmac.loc["Price in US Dollars", ("2016-01-01", "Brazil")]
```

# The .swaplevel() Method

# The .stack() Method

```python
world = pd.read_csv("worldstats.csv", 
                    index_col=["country", "year"]).sort_index()
world.head(3)
```

```python
world.stack().head()
```

```python
world.stack().to_frame().head()
```

# The .unstack() Method, Part 1

```python
world.head(3)
```

```python
s = world.stack()
s.head()
```

```python
s.unstack().head()
```

```python
s.unstack().unstack().head()
```

```python
s.unstack().unstack().unstack().head()
```

# .unstack Part 2

```python
world.head(3)
```

```python
s = world.stack()
s.head(3)
```

```python
s.unstack(2).head(3)
```

```python
s.unstack(1).head(3)
```

```python
s.unstack(0).head(3)
```

```python
s.unstack(-2).head(3)
```

```python
s.unstack("country").head(3)
```

# The .unstack() Method, Part 3

```python
s.head(3)
```

```python
s.unstack(["country", "year"]).head(3)
```

```python
s.unstack([-1, "year"]).head(3)
```

```python
s.unstack([1, 2]).head(3)
```

```python
s.unstack("year", fill_value=0).head()
```

# The .pivot() Method

```python
sales = pd.read_csv("salesmen.csv", parse_dates=["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales.pivot(index="Date", columns="Salesman", values="Revenue").head()
```

# The .pivor_table() Method

```python
foods = pd.read_csv("foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values="Spend", index="Gender", aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], 
                  columns="City", aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], 
                  columns=["Frequency", "City"], aggfunc="sum")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], 
                  columns="City", aggfunc="count")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], 
                  columns="City", aggfunc="max")
```

```python
foods.pivot_table(values="Spend", index=["Gender", "Item"], 
                  columns="City", aggfunc="min")
```

```python
pd.pivot_table(foods, values="Spend", index=["Gender", "Item"], 
                  columns="City", aggfunc="min")
```
