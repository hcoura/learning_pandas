# Filter df based on a condition

```python
df = pd.read_csv("employees.csv", parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
df[df["Gender"] == "Male"].head(3)
```

```python
mask = df["Team"] == "Finance"
df[mask].head(3)
```

```python
df[df["Senior Management"]].head(3)
```

```python
df[df["Salary"] > 110000].head(3)
```

# Filter with multiple conditions(AND)

```python
df.head(3)
```

```python
males = df["Gender"] == "Male"
finance = df["Team"] == "Finance"
df[males & finance].head(3)
```

```python
df[(df["Gender"] == "Male") & (df["Team"] == "Finance")].head(3)
```

```python
df[(df["Salary"] > 100000) & (df["Salary"] < 130000)].head(3)
```

# Filter with multiple conditions(OR)

```python
df.head(3)
```

```python
df[(df["Senior Management"] == False) | (df["Bonus %"] > 7)].head(3)
```

```python
# and or
m1 = df["Senior Management"] == False
m2 = df["Salary"] < 100000
m3 = df["Bonus %"] > 7
df[(m1 & m2) | m3].head(3)
```

# The .isin() Method

```python
df["Team"].isin(["Legal", "Sales", "Product"]).head(6)
```

```python
mask = df["Team"].isin(["Legal", "Sales", "Product"])
df[mask].head(3)
```

# .between()

```python
df.head(3)
```

```python
mask = df["Salary"].between(100000, 120000)

df[mask].head(6)
```

# .duplicated()

```python
fname = df["First Name"].sort_values()
fname.head()
```

```python
fname.duplicated().head()
```

```python
fname.duplicated(keep='last').head()
```

```python
fname.duplicated(keep=False).head()
```

```python
df[~df["First Name"].duplicated(keep=False)]
```

# .drop_duplicates()

```python
df.head()
```

```python
df.sort_values("First Name", inplace=True)
df.head()
```

```python
df.drop_duplicates().head()
```

```python
df.drop_duplicates("First Name").head()
```

```python
df.drop_duplicates("First Name", keep='last').head()
```

```python
df.drop_duplicates("First Name", keep=False).head()
```

```python
df.drop_duplicates(["First Name", "Team"], keep=False).head()
```
